```python
import numpy as np
import json
```

```python
with open("math_val.json") as f:
    data = json.load(f)
```

```python
num_correct = 0
num_sample = len(data)
for idx, sample in enumerate(data):
    random_choice = np.random.default_rng(seed=idx).choice(len(sample["choices"]), 1)
    if sample["choices"][random_choice[0]] == sample["answer"]:
        num_correct += 1

```

```python
print(f"Accuracy = {num_correct / num_sample}")
```
